#https://zim.vn/tieng-anh-chuyen-nganh-y

```python
url = 'https://zim.vn/tieng-anh-chuyen-nganh-y'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
rows = soup.find_all('tr')

for row in rows:
    row_text = ':'.join([tag.get_text(strip=True) for tag in row.find_all()])
    row.string = row_text

dataset = []

for row in rows:
    dataset.append(row.text)
```

```python
result = []

last_items_to_ignore = 17

for index, item in enumerate(dataset):
    if index < len(dataset) - last_items_to_ignore:
        # Tách thông tin từ mỗi phần tử trong mảng
        parts = item.split(':')
        # Lấy tên chuyên khoa và mô tả tiếng Việt, sau đó thêm vào danh sách kết quả
        result.append(parts[0] + ': ' + parts[-1])
    else:
        result.append(item)

for i in range(len(result[-17:])):
    item = result[-17 + i]
    parts = item.split(':')
    result[-17 + i] = f"{parts[2]}: {parts[4]}"

result = result[2:]
result = [item for item in result if "Nghĩa" not in item]
```

```python
dataset = []
for pair in result:
    parts = pair.split(':')
    en_word = parts[0].strip()
    vi_word = parts[1].strip()
    dataset.append({'en': en_word, 'vi': vi_word})

for item in dataset:
    my_data.append(item)

dataset
```

#https://langmaster.edu.vn/500-tu-vung-tieng-anh-chuyen-nganh-y-can-biet

```python
url = 'https://langmaster.edu.vn/500-tu-vung-tieng-anh-chuyen-nganh-y-can-biet'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
rows = soup.find_all(['li', 'b'])
newdata = []


for row in rows:
    # Tạo một danh sách để chứa nội dung của các thẻ <span> trong thẻ <li> hoặc nội dung của thẻ <b>
    contents = []
    for tag in row.find_all(['span', 'b']):  # Thêm thẻ b vào danh sách các thẻ cần tìm
        contents.append(tag.get_text())

    if contents:
        # Gộp nội dung của các thẻ <span> và/hoặc thẻ <b> lại với nhau
        merged_content = ' '.join(contents)
        # Thêm vào newdata
        newdata.append(merged_content)

if newdata:
    del newdata[0]
```

```python
for index, item in enumerate(newdata):
    newdata[index] = re.sub(r' –.*?:', ':', item)
```

```python
new_items = []

# Biểu thức chính quy để kiểm tra ký tự phiên âm
phonetic_pattern = re.compile(r'/[^\s]+/')

# Lặp qua từng mục trong danh sách và kiểm tra
for item in newdata:
    # Kiểm tra xem mục có chứa ký tự phiên âm không và không chứa ký tự '/'
    if not phonetic_pattern.search(item) and '/' not in item:
        # Nếu không có ký tự phiên âm và không chứa ký tự '/', thêm mục vào danh sách mới
        new_items.append(item)

del newdata
del new_items[-23:]
```

```python
dataset = []

for item in new_items:
    # Tách mỗi mục thành cặp từ tiếng Anh và tiếng Việt bằng cách tìm vị trí của dấu hai chấm
    colon_index = item.find(':')
    if colon_index != -1:
        # Lấy từ tiếng Anh từ đầu đến dấu hai chấm (bao gồm cả khoảng trắng ở đầu và cuối)
        en_word = item[:colon_index].strip()
        # Lấy từ tiếng Việt từ sau dấu hai chấm (bao gồm cả khoảng trắng ở đầu và cuối)
        vi_word = item[colon_index + 1:].strip()

        # Thêm cặp từ vào danh sách dataset dưới dạng từ điển
        dataset.append({'en': en_word, 'vi': vi_word})

for item in dataset:
    my_data.append(item)

dataset
```

##https://hinhanhykhoa.com/tieng-anh-y-khoa/song-ngu-hoi-dap-ve-benh-dau-mua-khi-monkeypox.html

```python
url = 'https://hinhanhykhoa.com/tieng-anh-y-khoa/song-ngu-hoi-dap-ve-benh-dau-mua-khi-monkeypox.html'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
spoilers = soup.find_all(class_="su-spoiler su-spoiler-style-default su-spoiler-icon-plus su-spoiler-closed")

dataset = []

for spoiler in spoilers:
    title = spoiler.find(class_="su-spoiler-title").text.strip()
    title_parts = title.split('\n')
    # Tạo từ điển cho mỗi phần tử và thêm vào danh sách dataset
    data = {'en': title_parts[1].strip(), 'vi': title_parts[0].strip()}
    dataset.append(data)

    content = spoiler.find(class_="su-spoiler-content").text.strip()

    sentences = re.split(r'(?<=[.!?])', content)
    sentences = [sentence for sentence in sentences if sentence.strip()]
    if len(sentences) % 2 == 0:
      temp = int(len(sentences)/2)

      for i in range(0, temp-1):
          en = sentences[i+temp]
          vi = sentences[i]
          dataset.append({'en': en, 'vi': vi})

```

```python
for item in dataset:
    my_data.append(item)
dataset
```

##https://hinhanhykhoa.com/kien-thuc-y-khoa/song-ngu-chan-doan-hinh-anh-mach-nao-o-benh-nhan-dot-quy-cap.html

```python
url = 'https://hinhanhykhoa.com/kien-thuc-y-khoa/song-ngu-chan-doan-hinh-anh-mach-nao-o-benh-nhan-dot-quy-cap.html'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
data = []

result = soup.find_all(class_="one_half")
for element in result:
    data.append(element.get_text().strip())

filtered_data = []
for item in data:
    if "nguồn: hinhanhykhoa.com" in item:
        item = item.replace("nguồn: hinhanhykhoa.com", "")  # Remove the specified string
    if "hinhanhykhoa.com" in item:
        item = item.replace("hinhanhykhoa.com", "")  # Remove the specified string
    filtered_data.append(item)
```

```python
dataset = []

leng = int(len(filtered_data)/2)

for i in range(0,leng):
  english_text = filtered_data[2*i]
  vietnamese_text = filtered_data[2*i+1]

  # Tách câu trong đoạn văn tiếng Anh
  english_sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', english_text)

  # Tách câu trong đoạn văn tiếng Việt
  vietnamese_sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', vietnamese_text)

  # Kiểm tra xem số lượng câu trong hai ngôn ngữ có bằng nhau không
  if int(len(english_sentences)) == int(len(vietnamese_sentences)):

      for i in range(len(english_sentences)):
          pair = {'en': english_sentences[i], 'vi': vietnamese_sentences[i]}
          dataset.append(pair)

  else:
      print("Số lượng câu trong hai ngôn ngữ không bằng nhau.")
```

```python
for item in dataset:
    my_data.append(item)
dataset
```

##https://hinhanhykhoa.com/tieng-anh-y-khoa/thuat-ngu-y-khoa-anh-viet-thuong-gap-az.html

```python
url = 'https://hinhanhykhoa.com/tieng-anh-y-khoa/thuat-ngu-y-khoa-anh-viet-thuong-gap-az.html'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
data = soup.find_all('p')
data = data[4:-2]
```

```python
new_data = []

for paragraph in data:
    new_data.append(paragraph.text)
```

```python
items_to_remove = []

for item in new_data:
    if '–' not in item:
        items_to_remove.append(item)

# Xóa các mục không chứa dấu '-' khỏi mảng
for item in items_to_remove:
    new_data.remove(item)
```

```python
dataset = []

for item in new_data:
    parts = item.split(' – ')
    if len(parts) == 2:
        en, vi = parts
        dataset.append({'en': en.strip(), 'vi': vi.strip()})
```

```python
for item in dataset:
    my_data.append(item)
dataset
```

##https://hinhanhykhoa.com/kien-thuc-y-khoa/song-ngu-chup-cat-lop-vi-tinh-co-tiem-thuoc-tuong-phan-va-cac-protocol.html

```python
url = 'https://hinhanhykhoa.com/kien-thuc-y-khoa/song-ngu-chup-cat-lop-vi-tinh-co-tiem-thuoc-tuong-phan-va-cac-protocol.html'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
data = []

result = soup.find_all(class_="one_half")
for element in result:
    text = element.get_text().strip()
    if text:  # Check if the string is not empty
        data.append(text)

data = [x.replace('\n', '') for x in data]
for sentence in data[:]:
    if "Thuyên tắc phổiNgực:" in sentence:
        data.remove(sentence)
data = data[:-18]

# Remove empty strings ('')
data = [x for x in data if x]

# Now, 'data' contains the desired list with empty strings removed
```

```python
num_sentences = len(data)
leng = int(num_sentences/2)
dataset = []
for i in range(0, leng):
  dataset.append({'en': data[2 * i], 'vi': data[2 * i + 1]})
```

```python
for item in dataset:
    my_data.append(item)
dataset
```
